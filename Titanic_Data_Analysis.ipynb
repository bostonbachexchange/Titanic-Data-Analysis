```python
import seaborn as sns
titanic_data = sns.load_dataset("titanic")
```

```python
titanic_data.head()
```

```python
titanic_data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
sns.histplot(titanic_data['age'].dropna(), kde=False)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()

```

```python
# Set the style of the seaborn plots
sns.set(style="whitegrid")

# Create a count plot to visualize survival by gender
sns.countplot(x='sex', hue='survived', data=titanic_data)

# Add labels and title to the plot
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Survival Count by Gender')

# Add a legend
plt.legend(['Not Survived', 'Survived'])

# Show the plot
plt.show()
```

```python
sns.histplot(titanic_data['fare'].dropna(), kde=False)
plt.xlabel('fare')
plt.ylabel('Count')
plt.title('Distribution of fare')
plt.show()
```

```python
# Set the style of the seaborn plots
sns.set(style="whitegrid")

# Create a count plot to visualize survival by gender
sns.countplot(x='class', hue='survived', data=titanic_data)

# Add labels and title to the plot
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Survival Count by Gender')

# Add a legend
plt.legend(['Not Survived', 'Survived'])

# Show the plot
plt.show()
```

```python
# Create a count plot to visualize survival by embarkation port
sns.countplot(x='embarked', hue='survived', data=titanic_data)

# Add labels and title to the plot
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survival Count by Embarkation Port')

# Add a legend
plt.legend(['Not Survived', 'Survived'])

# Show the plot
plt.show()
```

```python
# Create a scatter plot with age, fare, survival status, and passenger class
sns.scatterplot(x='age', y='fare', hue='survived', style='pclass', data=titanic_data)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare (Survival Status and Passenger Class)')

# Show legend
plt.legend(title='Survival', loc='upper right')

# Show the plot
plt.show()

```
